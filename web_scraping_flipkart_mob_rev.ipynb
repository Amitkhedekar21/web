```python
import numpy as np
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# url='https://www.flipkart.com/search?q=mi%20mobile%20phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'+str(i)
```

```python
mobile_name=[]
mobile_price=[]
mobile_rating=[]
pgno=int(input('Enter no of page: '))
for i in range(1,pgno+1):
    url="https://www.flipkart.com/search?q=samsung+mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&as-pos=1&as-type=RECENT&suggestionId=samsung+mobiles%7CMobiles&requestId=844c0384-f2f4-4840-9148-c1dffd6409a3&as-searchtext=samsu"+str(i)
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    name=soup.find_all(attrs={'class':'_4rR01T'})
    price=soup.find_all('div',{'class':"_30jeq3 _1_WHN1"})
    rating=soup.find_all('div',{'class':"_3LWZlK"})
    print("phone in page",str(i))
    for i in name:
        mobile_name.append(i.text)
    for i in price:
        mobile_price.append(i.text)
    for i in rating:
        mobile_rating.append(i.text)    
    

    
    
```

```python
print('mobile_name')        
print(mobile_name)
print('mobile_price')
print(mobile_price)
print('mobile_rating')
print(mobile_rating)
```
